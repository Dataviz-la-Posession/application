### Fusion du fichier CSV + GEOJSON

```python
# Charger le fichier CSV
csv_file_path = './data/csv_folder/rendu.csv'
csv_data = pd.read_csv(csv_file_path)

# Convertir "Code commune" en chaîne de caractères si nécessaire
if csv_data['Code commune'].dtype != 'object':
    csv_data['Code commune'] = csv_data['Code commune'].astype(str)

# Charger le fichier GeoJSON
geojson_file_path = './data/geojson/communes-974-la-reunion.geojson'
geojson_data = gpd.read_file(geojson_file_path)

# Convertir "code" du GeoJSON en chaîne de caractères si nécessaire
if geojson_data['code'].dtype != 'object':
    geojson_data['code'] = geojson_data['code'].astype(str)

# Fusionner les données sur "Code commune" et "code"
merged_data = geojson_data.merge(csv_data, left_on='code', right_on='Code commune')
```
